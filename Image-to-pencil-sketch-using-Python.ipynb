# Importing All the required libraries

```python
from PIL import Image
from IPython.display import display
import cv2
```

# Loading the image

```python
image = cv2.imread('Flag.jpg', 1)
display(Image.fromarray(image))
```

# Resizing the image

```python
image = cv2.resize(image, (640, 400))
display(Image.fromarray(image))
```

# Converting an image into rgb_scale image

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img_rgb))
```

# Converting an image into gray_scale image

```python
img_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(img_grey))
```

# Inverting the Image

```python
img_inv = cv2.bitwise_not(img_grey)
display(Image.fromarray(img_inv))
```

# Smoothing the image

```python
img_smooth = cv2.GaussianBlur(img_inv, (5, 5), 0, 0)
display(Image.fromarray(img_smooth))
```

# Inverting the image to get final sketch

```python
img_smooth_inv = 255 - img_smooth
img_final = cv2.divide(img_grey, img_smooth_inv, scale=256.0)
display(Image.fromarray(img_final))
```

# Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(image))
display(Image.fromarray(img_final))
```
